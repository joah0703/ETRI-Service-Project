```python
import os
import numpy as np
import pandas as pd
from tqdm import tqdm
import time
os.chdir(r'F:\etri\P27_구간찾기\sas')

data = pd.read_csv('P27_400_sas_1.csv')
del data['obs']

temp = np.zeros((data.shape[0],6))
confidence = np.zeros((data.shape[0],6)) #count 값 저장

def avg(a,b):       
    return (a+b)/2 
```

```python
for i in range(0,data.shape[0]):
    temp[i,0] = avg(data.iloc[i,1],data.iloc[i,0]) #up
    temp[i,1] = avg(data.iloc[i,3],data.iloc[i,2]) #down
    temp[i,2] = avg(data.iloc[i,5],data.iloc[i,4]) #left
    temp[i,3] = avg(data.iloc[i,7],data.iloc[i,6]) #right
    temp[i,4] = avg(data.iloc[i,9],data.iloc[i,8]) #eye 
    temp[i,5] = avg(data.iloc[i,11],data.iloc[i,10]) #mouth 
```

```python
for k in tqdm(range(0,temp.shape[1])): #신뢰구간     
    for j in tqdm(range(0,temp.shape[0])): 
        time.sleep(0.1)
        count = 0
        index = 0
        for x in range(0,temp.shape[0]):
            if temp[j,k] >= data.iloc[x,2*k]:
                if temp[j,k] <= data.iloc[x,2*k+1]:
                    count = count + 1
        
        confidence[j,k] = count
```

```python
confidence = pd.DataFrame(confidence)
sum_con = confidence.sum(axis=1)
confidence['sum'] = sum_con
```

```python
max_con = []
for x in range(0,len(confidence)):
    if confidence.loc[x,'sum'] == max(confidence['sum']):
        max_con.append(x)

data.iloc[max_con].sum(axis=0)/len(max_con)
```

```python
confidence.to_csv('P27_400_1_coverage.csv')
```
